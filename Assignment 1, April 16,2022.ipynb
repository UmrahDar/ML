## Write 5 different functions and call them in one program.

```python
def addition(a,b):
    return(a+b)
def prd(a,b):
    return(a*b)
def difference(a,b):
    return(a-b)
def division (a,b):
    return(a/b)
def modulus(a,b):
    return(a%b)
a=int(input("Enter first number:"))
b=int(input("Enter second number:"))
s= addition(int(a),b)
d= difference(a,b)
p= prd(a,b)
r= division(a,b)
m= modulus(a,b)

print("The sum is: ",s)
print("The product is: ",p)
print("The difference is: ",d)
print("The quotient is: ",r)
print("The remainder is: ",m)
```
